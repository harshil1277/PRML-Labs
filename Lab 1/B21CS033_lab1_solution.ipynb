```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression as LR
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.metrics import confusion_matrix as cm
from sklearn.metrics import precision_score as ps
from sklearn.metrics import recall_score as rs
from sklearn.metrics import f1_score as f1s
from sklearn.metrics import accuracy_score as acc
```

```python
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
```

```python
# Dataset Reference :- https://www.kaggle.com/uciml/breast-cancer-wisconsin-data

data = pd.read_csv("/content/drive/MyDrive/PRML lab 1 data /data.csv")
data
```

```python
# Note :- There are many existing Encoders for converting String to Numeric Labels, but for convenience, we used Pandas.

condition_M = data.diagnosis == "M"
condition_B = data.diagnosis == "B"

data.loc[condition_M,"diagnosis"]=0
data.loc[condition_B,"diagnosis"]=1

data
```

```python
Y = data.diagnosis.to_numpy().astype('int')                                     # Labels

X_data = data.drop(columns=["id","diagnosis","Unnamed: 32"])
X = X_data.to_numpy()                                                           # Input Features
```

```python
user_prompt = 0.3
user_enable = False

x_train,x_test,y_train,y_test = tts(X,Y,test_size=user_prompt,shuffle=user_enable)
```

```python
# Note :- Don't worry about the code snippet here, it is just to produce the predictions for the test data portion of each classifier

logistic_model = LR()
logistic_model.fit(x_train,y_train)
logistic_pred = logistic_model.predict(x_test)

decision_model = DTC()
decision_model.fit(x_train,y_train)
decision_pred = decision_model.predict(x_test)
```

## Confusion matrix

```python
# inbuilt_matrix_logistic = cm(y_test,logistic_pred)
# inbuilt_matrix_decision = cm(y_test,decision_pred)

# print("Confusion Matrix for Logistic Regression-based Predictions =>")
# print(inbuilt_matrix_logistic)
# print("Confusion Matrix for Decision Tree-based Predictions =>")
# print(inbuilt_matrix_decision)

def confusion_matrix(y_test,predicted):
  tp = 0 #true-positive
  tn = 0 #true-negative
  fp = 0 #false-positive
  fn = 0 #false-negative
  # 1 -- positive
  # 0 -- negative
  for i in range(len(y_test)):
    if(y_test[i]==1):
      if(predicted[i]==1):
        tp+=1
      else:
        fn+=1
    else:
      if(predicted[i]==1):
        fp+=1
      else:
        tn+=1
  #print(tp,tn,fp,fn)
  l=[tp,fp,fn,tn]
  l=np.array(l)
  l=np.reshape(l,(2,2))
  #print(l)
  return tp,fp,fn,tn
print("Confusion Matrix for Logistic Regression-based Predictions =>")  
print(confusion_matrix(y_test,logistic_pred))
print("Confusion Matrix for Decision Tree-based Predictions =>")
print(confusion_matrix(y_test,decision_pred))

# # print(y_test)
# # print(logistic_pred)
```

## Average Accuracy

```python
# inbuilt_acc_logistic = acc(y_test,logistic_pred)
# inbuilt_acc_decision = acc(y_test,decision_pred)

# print("Accuracy for Logistic Regression-based Predictions =>",str(inbuilt_acc_logistic*100)+"%")
# print("Accuracy for Decision Tree-based Predictions =>",str(inbuilt_acc_decision*100)+"%")

def avg_accuracy(y_test,predicted):
  tp,fp,fn,tn=confusion_matrix(y_test,predicted)
  return (tp+tn)/(tp+tn+fn+fp)*100
print("Accuracy for Logistic Regression-based Predictions =>")
print(avg_accuracy(y_test,logistic_pred))

print("Accuracy for Decision Tree-based Predictions =>")
print(avg_accuracy(y_test,decision_pred))  
```

## Precision

```python
# inbuilt_ps_logistic = ps(y_test,logistic_pred)
# inbuilt_ps_decision = ps(y_test,decision_pred)

# print("Precision for Logistic Regression-based Predictions =>",str(inbuilt_ps_logistic*100)+"%")
# print("Precision for Decision Tree-based Predictions =>",str(inbuilt_ps_decision*100)+"%")

def precision(y_test,predicted):
  tp,fp,fn,tn=confusion_matrix(y_test,predicted)
  return tp/(tp+fp)*100

print("Precision for Logistic Regression-based Predictions =>",precision(y_test,logistic_pred))  
print("Precision for Decision Tree-based Predictions =>",precision(y_test,decision_pred))    
```

## Recall

```python
# inbuilt_rs_logistic = rs(y_test,logistic_pred)
# inbuilt_rs_decision = rs(y_test,decision_pred)

# print("Recall for Logistic Regression-based Predictions =>",str(inbuilt_rs_logistic*100)+"%")
# print("Recall for Decision Tree-based Predictions =>",str(inbuilt_rs_decision*100)+"%")

def recall(y_test,predicted):
  tp,fp,fn,tn=confusion_matrix(y_test,predicted)
  return tp/(tp+fn)*100

print("Recall for Logistic Regression-based Predictions =>",recall(y_test,logistic_pred))
print("Recall for Decision Tree-based Predictions =>", recall(y_test,decision_pred))
```

## F-1 Score

```python
# inbuilt_f1s_logistic = f1s(y_test,logistic_pred)
# inbuilt_f1s_decision = f1s(y_test,decision_pred)

# print("F1-Score for Logistic Regression-based Predictions =>",str(inbuilt_f1s_logistic*100)+"%")
# print("F1-Score for Decision Tree-based Predictions =>",str(inbuilt_f1s_decision*100)+"%")

def f1_score(y_test,predicted):
  r=recall(y_test,predicted)
  p=precision(y_test,predicted)

  return 2/((1/r) + (1/p))

print("F1-Score for Logistic Regression-based Predictions =>",str(f1_score(y_test,logistic_pred))+"%")  
print("F1-Score for Decision Tree-based Predictions =>",str(f1_score(y_test,decision_pred))+"%")
```

## Class-Wise Accuracy

```python
def class_accuracy(y_test,predicted):
  tp,fp,fn,tn=confusion_matrix(y_test,predicted)
  print("positive class accuracy =", tp/(tp+fn)*100)
  print("negative class accuracy =", tn/(fp+tn)*100)

print("Class Accuricies for Logistic Regression-based Predictions =>")
class_accuracy(y_test,logistic_pred)  
print("Class Accuracies for Decision Tree-based Predictions =>") 
class_accuracy(y_test,decision_pred)   
```

## Sensitivity

```python
def sensitivity(y_test,predicted):
  tp,fp,fn,tn=confusion_matrix(y_test,predicted)
  return tp/(tp+fn)*100

print("Sensitivity for Logistic Regression-based Predictions =>",sensitivity(y_test,logistic_pred))  
print("Sensitivity for Decision Tree-based Predictions =>",sensitivity(y_test,decision_pred))     
```

## Specificity

```python
def specificity(y_test,predicted):
  tp,fp,fn,tn=confusion_matrix(y_test,predicted)
  return fn/(fn+tp)*100

print("specificity for Logistic Regression-based Predictions =>",specificity(y_test,logistic_pred))  
print("specificity for Decision Tree-based Predictions =>",specificity(y_test,decision_pred)) 
```
